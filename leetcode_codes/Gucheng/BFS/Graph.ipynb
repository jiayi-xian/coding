  + https://leetcode.cn/problems/course-schedule-ii/description/
  + description
    + 现在你总共有 numCourses 门课需要选，记为 0 到 numCourses - 1。给你一个数组 prerequisites ，其中 prerequisites[i] = [ai, bi] ，表示在选修课程 ai 前 必须 先选修 bi 。
    + 例如，想要学习课程 0 ，你需要先完成课程 1 ，我们用一个匹配来表示：[0,1] 。返回你为了学完所有课程所安排的学习顺序。可能会有多个正确的顺序，你只要返回 任意一种 就可以了。如果不可能完成所有课程，返回 一个空数组。

```python
# 207 Course Schedule
class Solution:
    def canFinish_1(self, numCourses: int, prerequisites: List[List[int]]) -> bool:

        
        indegree = [0 for _ in range(numCourses)]
        graph = defaultdict(set)

        if len(prerequisites) == 0: return True

        for c, prec in prerequisites:
            graph[prec].add(c)
            indegree[c] += 1
            if c == prec: return False

        # check if there is a cycle
        # ind1 = [ x for x in indegree if indegree[x]==0] # all the courses which don't need pre courses
        ind1 = [ x for x in range(numCourses) if indegree[x]==0]
        q = deque(ind1)
        visited = set(ind1)
        cnt = len(ind1)

        while q:
            prec = q.popleft()
            for c in graph[prec]:
                indegree[c] -= 1
                # if c not in visited and indegree[c] == 0:
                if indegree[c] == 0:
                    q.append(c)
                    indegree[c] = -1
                    cnt += 1
                    # visited.add(c) # 漏了
        
        return cnt == numCourses

    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:

        graph = {i:set() for i in range(numCourses)}

        for c, prec in prerequisites:
            graph[prec].add(c)

        visited = [0] * numCourses
        def dfs(idx):
            # assign 2 if we are visiting a node, 1 if finish visiting, 0: unvisited
            # a node represent a course, check if there exists a cycle within the graph of nodes (courses)
            if visited[idx] == 2: # we are visiting a node which are visiting
                return False
            
            if visited[idx] == 1: # we are visiting a node which has been visited
                return True

            visited[idx] = 2
            for c in graph[idx]:
                if dfs(c) is False:
                    return False
            
            visited[idx] = 1
            return True
        
        for idx in range(numCourses):
            if visited[idx] != 1:
                if dfs(idx) is False:
                    return False
        
        return True
```
