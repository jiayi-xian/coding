什么是双指针算法?

双指针算法是一种常见的算法, 核心理念是使用两个指针, 分别指向不同的元素, 通过对比指向的元素来协同完成任务。

若两个指针指向同一数组，遍历方向相同且不会相交，则称为滑动窗口，用于解决区间搜索问题。

还有快慢指针，可以用来判断链表环路问题。
