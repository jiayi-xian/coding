#### 1675. 数组的最小偏移量
给你一个由 n 个正整数组成的数组 nums 。

你可以对数组的任意元素执行任意次数的两类操作：

如果元素是 偶数 ，除以 2
例如，如果数组是 [1,2,3,4] ，那么你可以对最后一个元素执行此操作，使其变成 [1,2,3,2]
如果元素是 奇数 ，乘上 2
例如，如果数组是 [1,2,3,4] ，那么你可以对第一个元素执行此操作，使其变成 [2,2,3,4]
数组的 偏移量 是数组中任意两个元素之间的 最大差值 。

返回数组在执行某些操作之后可以拥有的 最小偏移量。

from typing import List

class Solution:
    def minimumDeviation(self, nums: List[int]) -> int:
        p_max = 1
        # 先把每个数都变成奇数 使之不能再变小 找到p_max
        for a in nums:
            while a % 2 == 0:
                a //= 2
            p_max = max(p_max, a)
        
        upper = []
        min_val = p_max
        for a in nums:
            if a % 2 == 1: # 奇数 *2 这个奇数一定小于p_max. a*=2 之后也不一定能大于p_max
                a *= 2
            while a >= 2 * p_max: # 这个a一定是偶数(因为p_max是最大的奇数) 可以写成p*2^R -> p*2^r < 2*p_max
                a //= 2
            if a >= p_max:
                # 这个a可能来源于第一步 a*=2的乘以2的奇数 也可能来源于第二步 变小的偶数 
                # 也可以是没有经过1，2步变化的偶数a。但a一定满足p_max <= a < 2*p_max
                # 总而言之，upper里面全是偶数 并且在(p_max, 2*p_max]范围内
                upper.append(a) 
            min_val = min(min_val, a) 
            # min_val来自p_max或者最小的偶数 这个偶数可能是原生的 也可能是第一步从更小的奇数得到的
        
        upper.sort()
        ans = upper[-1] - min_val
        for i in range(len(upper) - 1, -1, -1):
            if upper[i] > p_max:
                min_val = min(min_val, upper[i] // 2)
                if i > 0:
                    ans = min(ans, upper[i - 1] - min_val)
        
        return ans
