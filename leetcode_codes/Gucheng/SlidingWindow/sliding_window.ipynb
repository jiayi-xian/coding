# sliding windows
maintain一个sliding window，使得window里面的元素符合题目的特点。一般先移动右端点，使得window里面的元素符合题目的要求，然后移动左端点直到不满足条件。求window相关的最值

#### 3 Longest Substring Without Repeating Characters
Given 一个字符串s 找出最长子字符串，使得子字符串中没有重复的字符
```
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        l, res = 0, 0
        cnt = defaultdict(int)
        for i in range(len(s)):
            c = s[i]
            cnt[c] += 1
            while cnt[c] > 1:
                lc = s[l]
                cnt[lc] -= 1
                l += 1
            res = max(res, i-l+1)
        return res
```
#### 159 Longest Substring with At Most Two Distinct Characters
Given 一个字符串，返回最长子字符串（长度）使得子字符串中最多只有两个不同的字符
```
class Solution:
    def lengthOfLongestSubstringTwoDistinct(self, s: str) -> int:     

        cnt = {}
        l = 0
        k = 2
        maxL = 0
        for i in range(len(s)):
            c = s[i]
            if c not in cnt:
                cnt[c] = 0
            cnt[c] += 1
            while len(cnt) > k:
                lc = s[l]
                cnt[lc] -= 1
                if cnt[lc] == 0:
                    del cnt[lc]
                l += 1
            maxL = max(maxL, i - l+1)
        
        return maxL
```

#### 395 Longest Substring with At Least K Repeating Characters
395 Given 一个字符串 $s$ and 一个整数 $k$, return the length of the longest substring of $s$ such that the frequency of each character in this substring is greater than or equal to $k$. 求最长子字符串(返回最大长度)使得每一个字符在这个字符串内至少有k个
直接使用模版的话 无法确定window往前扩张还是收缩
如果返回的是数量的话 （at Least K repeating Characters 的substring 的数目）则有
exact(unique_chars, k) = atmost(unique_chars, k) - atmost(unique_chars-1, k)
exact(unique_chars, k)：满足条件的substring有正好unique_chars数目的chars 每个repeat至少k次
atmost(unique_chars, k)： 满足条件的substring有最多有unique_chars数目的chars且每个repeat至少k次
sum(exact(k) + ... + exact(len(s))) = atmost(len(s)) - atmost(k-1)

考虑 substring with at least k repeated characters and with x unique characters
```
    def longestSubstring(self, s: str, k: int) -> int:
```

#### 424 Longest Repeating Character Replacement 最长重复字符串置换
Given 一个字符串，规定最多可以变 $\mathrm{k}$ 个字符, 使得window里面全部都是同样的char。问最长的变化后的字符串是多长。
模板题直接上模板, 注意这里说了只有单词, 可以开 26 空间的 count array。
`i`前进 `for i in range(len(s))`
计算`cnt[c]`
计算 `maxC = max(cnt.values())` 
如果需要更改的字符串>k (`i - l + 1 - maxC > k`)
`cnt[lc] -= 1` (这一步一定要在`l`之前)
`l += 1`
最后计算`maxL`
```
def characterReplacement(s: str, k: int) -> int:

    cnt = defaultdict(int)
    i = 0
    maxC = 0
    maxChar = ""
    maxL = 0
    for j in range(len(s)):
        c = s[j]
        cnt[c] += 1
        if cnt[c] > maxC:
            maxC = cnt[c]
            maxChar = ""
        while j - i - maxC+1 > k: # O(1) max 26 letters
            cnt[s[i]] -= 1
            i += 1
    
        maxL = max(maxL, j-i+1)
    return maxL
```

#### 992 Subarrays with K Different Integers
Given一个整数数组和一个整数k，求“good”子数组的数目（一个子数组叫做“好的子数组”当且仅当该子数组有k个不同的元素）
If a subarray has $\mathrm{K}$ distinct elements, then total subarrays will be $n*(n+1) / 2$, which will include subarrays with $k-1, k-2, k-3 \ldots 1$ distinct elements.
套娃题
example: $[1,2,3] \mathrm{k}=1$
res: [1] [2] [3] [1,2] [2,3] [1,2,3]
res $=(1+3) * 3 / 2=6$ 高斯算法小于k都算
res $=1+2+3=6 \quad$ 等差数列求和公式
```
class Solution:
    def subarraysWithKDistinct(self, nums: List[int], k: int) -> int:
        A = nums

        def atMostK(K):
            """
            Given an array, return the subarray with at most K unique elements
            """
            res = 0
            i, j = 0, 0
            cnt = defaultdict(int)
            count = 0

            for j in range(len(A)):

                cnt[A[j]] += 1
                if cnt[A[j]] == 1: count += 1

                while count > K:
                    cnt[A[i]] -= 1
                    if cnt[A[i]] == 0:
                        count -= 1
                    i += 1
                
                res += j- i +1
            return res
        
        return atMostK(k) - atMostK(k-1)
```

#### Summary -- sliding window
1. Sliding window 套路模板时间复杂度一般为 $O(n)$
2. 一般string使用map作为window, 如果说明了只有小写字母也可以用int[26]
3. 多重限制条件的压轴题需要考虑是否为单调队列, 在另一节PPT有详解
4. 字母类还可以暴力尝试26个字母, 比如1个unique, 2个unique, 然后内部模板
5. Exact $(k)$ 可以转换为 atMost $(k)-\operatorname{atMost}(k-1)$

````python
#### 1248 Count Number of Nice Subarrays
"""
Given an array of integers nums and an integer $k$. A continuous subarray is called nice if there are $\mathrm{k}$ odd numbers on it.

Return the number of nice sub-arrays.
套娃题, 直接模板。
$\operatorname{exact}(k)$ 转化为 $\operatorname{atMost}(k)-\operatorname{atMost}(k-1)$
然后sliding window去找所有符合条件的 subarray, 再次高斯算法展开式求和
"""

def length_of_longest_substring_k_distinct(s, k):
    char_map = {}
    left = 0
    res = 0
    
    for i, cur in enumerate(s):
        char_map[cur] = char_map.get(cur, 0) + 1  # Add current character to the map
        
        while len(char_map) > k:  # Shrink the window from the left
            left_char = s[left]
            char_map[left_char] -= 1
            if char_map[left_char] == 0:
                del char_map[left_char]  # Remove the char from map if count goes to 0
            left += 1
        
        res = max(res, i - left + 1)  # Update result if we found a longer valid substring
    
    return res
````

````python
# longest substring with At least K repeating Characters
# template 只适合 distinct Character 所以转化为 max length of substrings with at least K repeating Characters with x unique characters
# when x from 1 to 26
def longest_substring(s, k):
    res = 0
    for unique in range(1, 27):  # Assuming we are dealing with the English alphabet
        char_map = {}
        left = 0
        valid_count = 0
        for i in range(len(s)):
            c = s[i]
            char_map[c] = char_map.get(c, 0) + 1
            if char_map[c] == k:
                valid_count += 1
            
            while len(char_map) > unique:
                left_char = s[left]
                char_map[left_char] -= 1
                if char_map[left_char] == 0:
                    del char_map[left_char]
                left += 1
            
            if len(char_map) == unique and valid_count == len(char_map):
                res = max(res, i - left + 1)
    return res

# Example usage
input_string = "aabbcc"
k = 2
print(longest_substring(input_string, k))  # Example output: This will depend on the input string and k
````
