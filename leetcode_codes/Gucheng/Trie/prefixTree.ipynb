```python
# 0000 0001 (27th bit)
"""
整合的思路
对所有输入的binary数建立trie
Give

"""
class TwoNumbersMaximumXor:
    def __init__(self):
        self.MAXN = 3000001
        self.tree = [[0] * 2 for _ in range(self.MAXN)]
        self.cnt = 1
        self.high = 31

    def findMaximumXOR1(self, nums):
        self.build(nums)
        ans = 0
        for num in nums:
            ans = max(ans, self.maxXor(num))
        self.clear()
        return ans

    def build(self, nums):
        self.cnt = 1
        max_num = max(nums)
        self.high = 31 - max_num.bit_length() + 1
        """
        31 - max_num.bit_length() + 1 这个表达式是用来确定
        一个整数 max_num 的最高有效位（也就是最左边的非零位）相对于某个位宽
        （在这里是32位，因为整数通常用32位表示）的位置。

        从左往右数，譬如0b101 则max_num.bit_length()为3 最高位是从左到右
        数31-3+1 = 29， 第29位（就是很右边）
        """
        for num in nums:
            self.insert(num)

    def insert(self, num):
        cur = 1
        for i in range(self.high, -1, -1):
            path = (num >> i) & 1
            """
            (num >> i) & 1  将第i位移到最右一位然后做且运算 来判断是0还是1 
            在计算机编程中经常用来检查整数 num 的二进制表示中第 i 位的值是0还是1
            """
            if self.tree[cur][path] == 0:
                self.cnt += 1
                self.tree[cur][path] = self.cnt
            cur = self.tree[cur][path]

    def maxXor(self, num):
        ans = 0
        cur = 1
        for i in range(self.high, -1, -1):
            status = (num >> i) & 1
            want = status ^ 1
            if self.tree[cur][want] == 0:
                want ^= 1
            ans |= (status ^ want) << i
            cur = self.tree[cur][want]
        return ans

    def clear(self):
        for i in range(1, self.cnt + 1):
            self.tree[i][0] = 0
            self.tree[i][1] = 0

    def findMaximumXOR2(self, nums):
        max_num = max(nums)
        ans = 0
        for i in range(31 - max_num.bit_length() + 1, -1, -1):
            better = ans | (1 << i)
            seen = set()
            for num in nums:
                num = (num >> i) << i
                seen.add(num)
                if better ^ num in seen:
                    ans = better
                    break
        return ans


# 解法2
"""
https://leetcode.cn/problems/maximum-xor-of-two-numbers-in-an-array
/solutions/2511670/python3javacgorust-yi-ti-yi-jie-0-1-zi-d-v28q/
"""
class Trie:
    __slots__ = ("children",)

    def __init__(self):
        self.children: List[Trie | None] = [None, None]

    def insert(self, x: int):
        node = self
        for i in range(30, -1, -1):
            v = x >> i & 1
            if node.children[v] is None:
                node.children[v] = Trie()
            node = node.children[v]

    def search(self, x: int) -> int:
        node = self
        ans = 0
        for i in range(30, -1, -1):
            v = x >> i & 1
            if node.children[v ^ 1]:
                ans |= 1 << i
                node = node.children[v ^ 1]
            else:
                node = node.children[v]
        return ans
```

```python
# 推荐 使用静态数组 这个和动态的区别在于 动态申请空间是每次测试一个实例就申请一个新的trie
# 在比赛中，空间使用是所有test cases 总用空间的大小
# 而使用静态数组 每次只建立一个trie （这不就是单例吗？）
class TrieTree:
    def __init__(self):
        self.MAXN = 150001
        self.tree = [[0] * 26 for _ in range(self.MAXN)]
        """
        
        """
        self.end = [0] * self.MAXN
        self.pass_ = [0] * self.MAXN  # 'pass' is a keyword in Python, so I'm using 'pass_' instead
        self.cnt = 1

    def build(self):
        self.cnt = 1

    def insert(self, word):
        cur = 1
        self.pass_[cur] += 1
        for char in word:
            path = ord(char) - ord('a')
            if self.tree[cur][path] == 0: # cur: the index of space required in total. 从1开始. path: ord(char) - ord("a") index of char in alphabet ordering 
                self.cnt += 1
                self.tree[cur][path] = self.cnt
                # 意思是 在该cur层中 我们有字母path 字母path(能到)新申请的空间: 下一层self.cnt
            cur = self.tree[cur][path]
            self.pass_[cur] += 1
        self.end[cur] += 1

    def search(self, word):
        cur = 1
        for char in word:
            path = ord(char) - ord('a')
            if self.tree[cur][path] == 0:
                return 0
            cur = self.tree[cur][path]
        return self.end[cur]

    def prefix_number(self, pre):
        """
        有多少记录的词以此pre为prefix， 注意这些词未必uniuqe
        如果想记载以当前pre为前缀的unique的词的数目
        """
        cur = 1
        for char in pre:
            path = ord(char) - ord('a')
            if self.tree[cur][path] == 0:
                return 0
            cur = self.tree[cur][path]
        return self.pass_[cur]

    def delete(self, word):
        if self.search(word) > 0:
            cur = 1
            for char in word:
                path = ord(char) - ord('a')
                if self.pass_[self.tree[cur][path]] - 1 == 0:
                    self.tree[cur][path] = 0
                    return
                self.pass_[self.tree[cur][path]] -= 1
                cur = self.tree[cur][path]
            self.end[cur] -= 1

    def clear(self):
        for i in range(1, self.cnt + 1):
            self.tree[i] = [0] * 26
            self.end[i] = 0
            self.pass_[i] = 0


def main():
    trie = TrieTree()
    m = int(input())
    trie.build()
    for _ in range(m):
        parts = input().split()
        op = int(parts[0])
        if op == 1:
            trie.insert(parts[1])
        elif op == 2:
            trie.delete(parts[1])
        elif op == 3:
            print("YES" if trie.search(parts[1]) > 0 else "NO")
        elif op == 4:
            print(trie.prefix_number(parts[1]))
    trie.clear()

if __name__ == "__main__":
    main()
```

```python
# 实现1 使用动态类
class TrieNode1:
    def __init__(self):
        self.pass_count = 0 # 有多少个词经过该节点
        self.end_count = 0 # 有多少个词以该节点为end
        self.nexts = [None] * 26 # 下一级的path

class Trie1:
    def __init__(self):
        self.root = TrieNode1() # 只需要记头节点就够了

    def insert(self, word):
        node = self.root
        node.pass_count += 1
        for ch in word:
            path = ord(ch) - ord('a')
            if node.nexts[path] is None:
                node.nexts[path] = TrieNode1()
            node = node.nexts[path]
            node.pass_count += 1
        node.end_count += 1

    def erase(self, word):
        if self.countWordsEqualTo(word) > 0:
            node = self.root
            node.pass_count -= 1
            for ch in word:
                path = ord(ch) - ord('a') # 找下一级节点
                if node.nexts[path].pass_count == 1:
                    node.nexts[path] = None
                    return # C++ C 需要手动释放节点 否则内存占用会越来越多
                node = node.nexts[path]
                node.pass_count -= 1 # 
            node.end_count -= 1 # pass还有 但endcount为0 此node不需要释放

    # 查询该单词在trie中出现过多少次
    def countWordsEqualTo(self, word):
        node = self.root
        for ch in word:
            path = ord(ch) - ord('a')
            if node.nexts[path] is None:
                return 0
            node = node.nexts[path]
        return node.end_count

    # 查询有多少个单词以此prefix作为前缀
    def countWordsStartingWith(self, prefix):
        node = self.root
        for ch in prefix:
            path = ord(ch) - ord('a')
            if node.nexts[path] is None:
                return 0
            node = node.nexts[path]
        return node.pass_count

class TrieNode2:
    def __init__(self):
        self.pass_count = 0
        self.end_count = 0
        self.nexts = {} # 应付所有可能的输入 65535 ascii 码

class Trie2:
    def __init__(self):
        self.root = TrieNode2()

    def insert(self, word):
        node = self.root
        node.pass_count += 1
        for ch in word:
            if ch not in node.nexts:
                node.nexts[ch] = TrieNode2()
            node = node.nexts[ch]
            node.pass_count += 1
        node.end_count += 1

    def erase(self, word):
        if self.countWordsEqualTo(word) > 0:
            node = self.root
            node.pass_count -= 1
            for ch in word:
                next_node = node.nexts[ch]
                if next_node.pass_count == 1:
                    del node.nexts[ch]
                    return
                node = next_node
                node.pass_count -= 1
            node.end_count -= 1

    def countWordsEqualTo(self, word):
        node = self.root
        for ch in word:
            if ch not in node.nexts:
                return 0
            node = node.nexts[ch]
        return node.end_count

    def countWordsStartingWith(self, prefix):
        node = self.root
        for ch in prefix:
            if ch not in node.nexts:
                return 0
            node = node.nexts[ch]
        return node.pass_count
```

```python
# 642 Design Search autocomplete

import heapq

class TrieNode:
    def __init__(self):
        self.frq = 0
        self.sentence = ""
        self.children = [None] * 27

class AutocompleteSystem:
    def __init__(self, sentences, times):
        self.root = TrieNode()
        self.curr_node = self.root
        self.pre = ""
        self.ret = []
        for sentence, time in zip(sentences, times):
            self.add_sentence(sentence, time)

    def add_sentence(self, sentence, time):
        node = self.root
        for ch in sentence:
            idx = 26 if ch == ' ' else ord(ch) - ord('a')
            if node.children[idx] is None:
                node.children[idx] = TrieNode()
            node = node.children[idx]
        node.frq = time
        node.sentence = sentence

    def dfs_find_sentences(self, node, heap):
        if node.frq > 0:
            heapq.heappush(heap, (-node.frq, node.sentence))
            if len(heap) > 3:
                heapq.heappop(heap)
        for child in node.children:
            if child is not None:
                self.dfs_find_sentences(child, heap)

    def input(self, c):
        if c == '#':
            self.add_sentence(self.pre, self.curr_node.frq + 1)
            self.curr_node = self.root
            self.pre = ""
            return []
        self.pre += c
        idx = 26 if c == ' ' else ord(c) - ord('a')
        if self.curr_node.children[idx] is None:
            self.curr_node.children[idx] = TrieNode()
        self.curr_node = self.curr_node.children[idx]

        heap = []
        self.dfs_find_sentences(self.curr_node, heap)
        heap.sort(reverse=True)  # To get the top 3 in descending order of frequency and lexicographic order
        return [sentence for frq, sentence in heap]

# Example usage
# autocomplete_system = AutocompleteSystem(["i love you", "island", "iroman", "i love leetcode"], [5, 3, 2, 2])
# print(autocomplete_system.input('i'))
# print(autocomplete_system.input(' '))
# print(autocomplete_system.input('a'))
# print(autocomplete_system.input('#'))
```
