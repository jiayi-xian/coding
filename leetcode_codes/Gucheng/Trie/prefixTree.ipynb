实现前缀树 Trie 类:
1. Trie() 初始化前缀树对象。
2. void insert(String word) 将字符串 word 插入前缀树中。
3. int search(String word) 返回前缀树中字符串 word 的实例个数。
4. int prefixNumber(String prefix) 返回前缀树中以 prefix 为前缀的字符串个数
5. void delete(String word) 从前缀树中移除字符串 word 。

分别用 类描述的方式、静态数组的方式 实现

```python
# 642 Design Search autocomplete

import heapq

class TrieNode:
    def __init__(self):
        self.frq = 0
        self.sentence = ""
        self.children = [None] * 27

class AutocompleteSystem:
    def __init__(self, sentences, times):
        self.root = TrieNode()
        self.curr_node = self.root
        self.pre = ""
        self.ret = []
        for sentence, time in zip(sentences, times):
            self.add_sentence(sentence, time)

    def add_sentence(self, sentence, time):
        node = self.root
        for ch in sentence:
            idx = 26 if ch == ' ' else ord(ch) - ord('a')
            if node.children[idx] is None:
                node.children[idx] = TrieNode()
            node = node.children[idx]
        node.frq = time
        node.sentence = sentence

    def dfs_find_sentences(self, node, heap):
        if node.frq > 0:
            heapq.heappush(heap, (-node.frq, node.sentence))
            if len(heap) > 3:
                heapq.heappop(heap)
        for child in node.children:
            if child is not None:
                self.dfs_find_sentences(child, heap)

    def input(self, c):
        if c == '#':
            self.add_sentence(self.pre, self.curr_node.frq + 1)
            self.curr_node = self.root
            self.pre = ""
            return []
        self.pre += c
        idx = 26 if c == ' ' else ord(c) - ord('a')
        if self.curr_node.children[idx] is None:
            self.curr_node.children[idx] = TrieNode()
        self.curr_node = self.curr_node.children[idx]

        heap = []
        self.dfs_find_sentences(self.curr_node, heap)
        heap.sort(reverse=True)  # To get the top 3 in descending order of frequency and lexicographic order
        return [sentence for frq, sentence in heap]

# Example usage
# autocomplete_system = AutocompleteSystem(["i love you", "island", "iroman", "i love leetcode"], [5, 3, 2, 2])
# print(autocomplete_system.input('i'))
# print(autocomplete_system.input(' '))
# print(autocomplete_system.input('a'))
# print(autocomplete_system.input('#'))
```
