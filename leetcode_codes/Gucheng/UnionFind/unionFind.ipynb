并查集 是一种树形的数据结构, 用于处理不交集的合并(union)及查询(find)问题。
If $\mathrm{m}$ operations, either Union or Find, are applied to $\mathrm{n}$ elements, the total run time is $O\left(m \log ^* n\right)$, where $\log ^*$ is the iterated logarithm.
Slightly larger than $O(m)$, it is Amortized $O(m)$ 类似(not same)平均, 实际为均摊

```python
# 547. Friend Circles (非常经典的union find)

class DSU:
    def __init__(self, n):
        self.p = [i for i in range(n)]

    def fp(self, v): # find parent
        if self.p[v] != v:
            self.p[v] = self.fp(self.p[v])
        return self.p[v]

    def union(self, u, v):
        pu = self.fp(u)
        pv = self.fp(v)
        self.p[pv] = pu

class Solution:
    def findCircleNum(self, isConnected: List[List[int]]) -> int:

        res = 0
        nrow = len(isConnected)
        dsu = DSU(nrow)

        for i in range(nrow):
            for j in range(i, nrow):
                if isConnected[i][j] == 1:
                    dsu.union(i, j)
        
        for i in range(nrow):
            dsu.fp(i)

        return len(set(dsu.p))

```

```python
# - 305. Number of Islands II
"""    
- Description
        - 给你一个大小为 m x n 的二进制网格 grid 。网格表示一个地图，其中，0 表示水，1 表示陆地。最初，grid 中的所有单元格都是水单元格（即，所有单元格都是 0）。
        - 可以通过执行 addLand 操作，将某个位置的水转换成陆地。给你一个数组 positions ，其中 positions[i] = [ri, ci] 是要执行第 i 次操作的位置 (ri, ci) 。
        - 返回一个整数数组 answer ，其中 answer[i] 是将单元格 (ri, ci) 转换为陆地后，地图中岛屿的数量。
        - 岛屿 的定义是被「水」包围的「陆地」，通过水平方向或者垂直方向上相邻的陆地连接而成。你可以假设地图网格的四边均被无边无际的「水」所包围。
    - union find vs BFS:
      - 这道题bfs肯定也可以做。每当insert一块island，check四周是否是visited的island，如果是，islands数保持不变，否则islands数加一
"""

class DSU:
    def __init__(self, N):
        self.parent = list(range(N))

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        self.parent[self.find(x)] = self.find(y)

class Solution:
    def numIslands2_0(self, m: int, n: int, positions: List[List[int]]) -> List[int]: 
        
        neis = lambda x, y: [(nx, ny) for nx, ny in [(x-1, y), (x+1, y), (x, y-1), (x, y+1)] if 0<=nx<m and 0<=ny<n]

        res = []
        dsu = DSU(m, n)
        P = set(tuple(p) for p in positions)
        islands = len(P)
        for r,c in positions:
            island_new = 1
            for (nr, nc) in neis(r, c):
                if (nr, nc) in P:
                    if (nr, nc) in visited:
                        island_new = 0
                    np2 = dsu.fp((nr,nc))
                    np1 = dsu.fp((r,c))
                    if np2 != np1:
                        dsu.union(np2, np1)
                        islands -= 1
        
        return islands
    def numIslands2(self, m, n, positions):
        dsu = DSU(m * n)
        island = [[False for _ in range(n)] for _ in range(m)] # visited matrix
        res = []
        count = 0
        dirs = [(-1, 0), (0, 1), (0, -1), (1, 0)]
        for cur in positions:
            if island[cur[0]][cur[1]]:
                res.append(count)
                continue
            island[cur[0]][cur[1]] = True
            count += 1
            for dir in dirs:
                x, y = cur[0] + dir[0], cur[1] + dir[1]
                if x < 0 or x >= m or y < 0 or y >= n or not island[x][y]: # island没有被visited过的不考虑 这是因为未visited的一定满足component1 != component2 但不一定导致count的减少 (count减少的原因是与已经visited过的island有连结)
                    continue
                component1 = dsu.find(cur[0] * n + cur[1])
                component2 = dsu.find(x * n + y)
                if component1 != component2:
                    dsu.union(component2, component1)
                    count -= 1
            res.append(count)
        return res

                
        
```

```python
# 128. Longest Consecutive Sequence

class DSU:
    def __init__(self, n= 0):
        self.p = {}

    def fp(self, v): # find parent
        if self.p[v] != v:
            self.p[v] = self.fp(self.p[v])
        return self.p[v]

    def union(self, u, v):
        pu = self.fp(u)
        pv = self.fp(v)
        self.p[pv] = pu


class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        # when we see a number, we check n+1 and n-1
        # if neighbor is visited, we mark it and union
        if nums == []:
            return 0
        dsu = DSU()
        for idx, n in enumerate(nums):
            if n not in dsu.p:
                dsu.p[n] = n
            if n-1 in dsu.p: dsu.union(dsu.fp(n), dsu.fp(n-1))
            if n+1 in dsu.p: dsu.union(dsu.fp(n+1), dsu.fp(n))
        
        cnt = {}
        for i in dsu.p.keys():
            key = dsu.fp(i)
            if key not in cnt:
                cnt[key] = 0
            cnt[key] += 1
        maxL = max(cnt.values())

        return maxL


```
