```python
"""
1095
先使用find peak 然后再 binary search
这道题充分说明了binary search 对于单调增和单调减是不一样的

"""
# """
# This is MountainArray's API interface.
# You should not implement it, or speculate about its implementation
# """
#class MountainArray:
#    def get(self, index: int) -> int:
#    def length(self) -> int:

class Solution:
    def findInMountainArray(self, target: int, mountain_arr: 'MountainArray') -> int:
        
        ma = mountain_arr
        peakIdx = self.getpeak(ma)
        lidx = self.bs(ma, target, 0, peakIdx)
        if peakIdx < ma.length()-1:
            ridx = self.bs2(ma, target, peakIdx+1, ma.length()-1)
        
        if lidx != -float("inf") and ma.get(lidx) == target: return lidx
        if ridx != -float("inf") and ma.get(ridx) == target: return ridx
        return -1



    def bs(self, ma, target, s, t):

        l, r = s, t
        if l == r:
            return l if ma.get(l) == target else -float("inf")
        
        while l<=r:

            m = l + (r-l)//2
            mval = ma.get(m)
            if mval == target:
                return m
            elif mval > target:
                r = m-1
            elif mval < target:
                l = m + 1
        
        return -float("inf")

    def bs2(self, ma, target, s, t):

        l, r = s, t
        if l == r:
            return l if ma.get(l) == target else -float("inf")
        
        while l<=r:

            m = l + (r-l)//2
            mval = ma.get(m)
            if mval == target:
                return m
            elif mval > target:
                l = m+1
            elif mval < target:
                r = m-1
        
        return -float("inf")

    def getpeak(self, ma):
        l, r = 0, ma.length()-1
        if l == r:
            return 0

        while l<r:
            m = l + (r-l)//2 + 1
            mval = ma.get(m)
            m1val = ma.get(m-1)

            if mval>m1val:
                l = m
            else:
                r = m-1
        return l
```

```python
"""
31 Next permutation
1. 我们希望下一个数 比当前数大, 这样才满足 "下一个排列" 的定义。因此只需要 将后面的「大数」与前面的「小数」交换, 就能得到一个更大的数。比如 123456 , 将 5 和 6 交换就能得到一个更大的数 123465 。
2. 我们还希望下一个数增加的幅度尽可能的小, 这样才满足“下一个排列与当前排列紧邻"的要求。为了满足这个要求, 我们需要:
a. 在 尽可能靠右的低位 进行交换, 需要 从后向前 查找
b. 将一个尽可能小的「大数」与前面的「小数」交换。比如 123465 , 下一个排列应该把 5 和 4 交换而不是把 6 和 4 交换
c. TODO 升序排列后面的很重要 不要忘记这一步 将「大数」换到前面后, 需要将「大数」后面的所有数 重置为升序, 升序排列就是最小的排列。以 123465 为例：首先按照上一步, 交换 5 和 4 , 得到 123564 ; 然后需要将 5 之后的数重置为升序, 得到 123546 。显然 123546 比 123564 更小, 123546 就是 123465 的下一个排列

以上就是求 “下一个排列” 的分析过程。

算法过程
标准的 “下一个排列” 算法可以描述为：
1. 从后向前 查找第一个相邻升序 的元素对 $(i, j)$, 满足 $A[i]<A[j]$ 。此时 $[j$, end $)$ 必然是降序
2. 在 $[j$, end 从后向前 查找第一个满足 $A[i]<A[k]$ 的 $k 。 A[i] 、 A[k]$ 分别就是上文所说的「小数」、「大数」
3. 将 $A[i]$ 与 $A[k]$ 交换
4. 可以断定这时 $[j$, end ) 必然是降序, 逆置 $[j$, end ), 使其升序
5. 如果在步骤 1 找不到符合的相邻元素对, 说明当前 [begin, end) 为一个降序顺序, 则直接跳到步骤 4

"""

class Solution:
    def nextPermutation(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """

        curMinMax = nums[-1]
        idx2swap = None
        
        for i in range(len(nums)-1, -1, -1):
            if i-1>=0:
                if nums[i] > nums[i-1]:
                    idx2swap = i-1
                    break # TODO 不要忘记break
        if idx2swap is None:
            nums.sort()
            return
        idx4swap = i
        curMinMax = nums[i] # TODO 要初始化
        for j in range(i+1, len(nums)):
            if nums[j] > nums[idx2swap]:
                if nums[j] < curMinMax:
                    curMinMax = nums[j] # TODO 不要忘记这一行
                    idx4swap = j
        nums[idx2swap], nums[idx4swap] = nums[idx4swap], nums[idx2swap]
        nums[idx2swap+1:] = sorted(nums[idx2swap+1:])
        return nums


```

```python
"""
670 maximum swap

给定一个非负整数，你至多可以交换一次数字中的任意两位。返回你能得到的最大值。

思路：
optimal
使用maxJ保存从右往左的最大index
每遍历一个数 将它与maxJ对应的数比较 如果小于maxJ对应的数 设交换的两数为p,q p对应当前数 q对应maxJ对应的数
从左往右 对每个数字 从右往左找到比它大的数字 保留数值最大且index最大的一个 （greedy 就是这两个数字要交换）。如果能找到 则交换 不能找到则返回原来的数
"""

"""
suboptimal
从左往右 对每个数字 从右往左找到比它大的数字 保留数值最大且index最大的一个 （greedy 就是这两个数字要交换）。如果能找到 则交换 不能找到则返回原来的数

"""

class Solution:
    def maximumSwap(self, num: int) -> int:
        s = str(num)
        max_idx = len(s) - 1
        p = q = -1
        for i in range(len(s) - 2, -1, -1):
            if s[i] > s[max_idx]:  # s[i] 是目前最大数字
                max_idx = i
            elif s[i] < s[max_idx]:  # s[i] 右边有比它大的
                p, q = i, max_idx  # 更新 p 和 q
        if p == -1:  # 这意味着 s 是降序的
            return num
        s = list(s)
        s[p], s[q] = s[q], s[p]  # 交换 s[p] 和 s[q]
        return int(''.join(s))

    def maximumSwap_suboptimal(self, num: int) -> int:

        maxE = 0
        maxJ = -1
        s = [int(c) for c in str(num)]
        flag = 0
        for i in range(len(s)):
            maxE = -float("inf")
            cur = s[i]
            for j in range(len(s)-1, i-1, -1):
                if int(s[j]) > maxE:
                    maxE  = max(maxE, int(s[j]))
                    maxJ = j
            if maxE > int(cur): 
                flag = 1
                break
        
        if not flag:
            return num
        else:
            s[i], s[maxJ] = s[maxJ], s[i]
            return int("".join(map(str, s)))



        
```

```python
"""
1060 Missing Element in Sorted Array


"""

class Solution:
    def missingElement(self, nums: List[int], k: int) -> int:

        """
        [4,7,9,10] - [1,2,3,4] -> 3, 5, 6, 6
        [4,7,9,10] - [4,5,6,7] -> 0, 2, 3, 3

        """
        if not nums: return k
        #if k < nums[0] : return k

        nums.append(nums[-1]+2*k)

        K = k
        l, r = 0, len(nums)-1

        # find the rightmost d in diff such that d < k
        while l<r:

            m = l + (r-l) //2 + 1
            # if diff[m] < k
            if nums[m] - (m+nums[0]) < k:
                l = m
            else:
                r = m-1
        
        # l: expected index found
        return nums[l] + k - (nums[l] - (l+nums[0])) # k+l+nums[0]
```

```python
"""
1539 the kth missing pos integer
给你一个 严格升序排列 的正整数数组 arr 和一个整数 k 。
请你找到这个数组里第 k 个缺失的正整数。

注意
edge case：
空数组
（数组为1）
missing integer不在数组中 而在数组之后 或者之前

增加pre=0
公式：res -= (cur - pre -1)
pre + res (或者 cur + res)
示例 1：

输入：arr = [2,3,4,7,11], k = 5
输出：9
解释：缺失的正整数包括 [1,5,6,8,9,10,12,13,...] 。第 5 个缺失的正整数为 9 。
"""
class Solution:
    def findKthPositive(self, arr: List[int], k: int) -> int:   
        if len(arr) == 0:
            return k

        res = k
        pre = 0
        for idx, n in enumerate(arr):
            if res - (n - pre -1) <= 0:
                return pre + res #小心写错成 n+pre
            else:
                res -= (n - pre -1) # 每一步 update了k之后 转化成找从当前数起missing的第k个正整数
            pre = n
        
        return n + res
```
