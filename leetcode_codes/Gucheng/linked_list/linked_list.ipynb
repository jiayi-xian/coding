````python
#### 1171 Remove Zero Sum Consecutive Nodes from Linked List
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeZeroSumSublists(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # create a dict called preSum to store prefix Sum and corresponeidng nodes. When we encounter the same prefix Sum, we delete the nodes between them node1.next = node2.next

        prefixSum = {}

        pre = ListNode(0)
        dummy = ListNode(0)
        dummy.next = head
        pre.next = head
        cur = head

        prefixSum[0] = dummy
        ps = 0
        while cur:
            ps += cur.val
            prefixSum[ps] = cur
            cur = cur.next
        ps = 0
        cur = dummy
        while cur:
            ps += cur.val
            cur.next = prefixSum[ps].next
            cur = cur.next

        return dummy.next


        
````

````python
#### Remove Nth Node From End of List
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # the right pointer starting from head, moves for n steps and ends at n+1 node
        # while right arrives at None, it moves N-n steps, so left moves N-n steps too.
        # left starting from dummy, after moving for N-n steps, it arrives at the N-n node
        # note that the n node from end  is the N-n+1 node from starts, so 
        # left.next = left.next.next removes the required node
        # return dummy.next
        dummy = ListNode(0, head)
        left = dummy
        right = head

        for i in range(n):
            right = right.next
        
        while right:
            left = left.next
            right = right.next
        
        left.next = left.next.next
    
        return dummy.next

    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # the last n-th node is the N-n+1-the node
        # first let p2 starting from head, moving n-1 step, it will arrive at the n-th node
        # then let p1 and pre staring from head, moving with p2 until p2 ends at None. While p2 ending at None, it bypass N-n nodes and moves N-n+1 steps, at the same time p1 arrives at N-n+2 -th node, we adjust it a bit, move p2 until p2.next is None. Then p1 arrives at N-n+1 - node, which is the n th node from end

        p1, p2 = head, head
        pre = ListNode(0)
        pre.next = p1
        p0 = pre

        for _ in range(n-1):
            if p2 is None: return head
            p2 = p2.next

        while p2.next: # if p2.next is None from beginning, p1 is at the n-th node from end
            p1 = p1.next
            pre = pre.next
            p2 = p2.next

        pre.next = p1.next
        return p0.next
````

````python
#### 83. Remove Duplicates from Sorted List
"""
Given the head of a sorted linked list, delete all duplicates such that each element appears 
only once. Return the linked list sorted as well.
"""
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates0(self, head: Optional[ListNode]) -> Optional[ListNode]:

        #dummy = ListNode(0)
        pre, cur = head, head.next if head else None

        while cur:
            if pre.val == cur.val:
                pre.next = cur.next
                cur = cur.next
                continue
            pre, cur = cur, cur.next
        
        return head

    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        def helper(node, val):

            cur = node
            if not cur: return None
            if cur.val == val:
                return helper(cur.next, val)
            else:
                cur.next = helper(cur.next, cur.val)
                return cur
        
        return helper(head, head.val+1) if head else None
````

````python
#### 206 reverse linked nodes

"""
1. pre, cur
2. nex 存储cur.next (下一时刻的cur)
3. reverse：update cur.next 为pre
4. move
"""

def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    pre, cur= None, head # 1
    while cur:
        nex = cur.next # 2
        cur.next = pre # 3
        pre, cur = cur, nex # 4
    
    return pre
````

#### 25 Reverse Nodes in k-Group
Given一个linked list 和一个正整数 k， 按照k组linked nodes进行reverse。如果最后剩下的nodes不足k个 则不进行reverse
Input: head = [1,2,3,4,5], k = 2
Output: [2,1 ｜,4,3｜,5]

````python
# k个nodes reverse则是进行k-1次reverse



# step 1 check if there is enough nodes to perform reverse operations
cur, pre = head, dummy

p1 = dummy # p1.next = ghead = last node reversed = cur
end = cur # node 1 end is the new p1 for the next k group  p1 = end
K = k
while K-1 and cur:
    pre, cur = cur, cur.next
    K -= 1
if K-1 == 0 and not cur:
    return
else:
    pre, cur, nex = p1, end, end.next
    # perform reverse operation
    for i in range(k):
        pre, cur, nex = cur, nex, nex.next if nex else None
        cur.next = pre
    p1.next = cur
    p1 = end
    p1.next = nex # 2->1->3->4 cur = 2
    pre, cur, nex = p1, nex, nex.next if nex else None
    end = cur
````

````python

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeZeroSumSublists(head):
    prefix = 0
    dummy = ListNode(0)
    dummy.next = head
    prefix_sum_map = {}
    prefix_sum_map[0] = dummy
    
    # First pass to calculate prefix sums
    current = dummy
    while current:
        prefix += current.val
        prefix_sum_map[prefix] = current
        current = current.next

    # Second pass to remove zero sum sublists
    prefix = 0
    current = dummy
    while current:
        prefix += current.val
        current.next = prefix_sum_map[prefix].next
        current = current.next
    
    return dummy.next
````

````python
#### 160. Intersection of Two Linked Lists
class Solution:
    def getIntersectionNode0(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        
        curA = headA
        curB = headB
        la , lb = 0, 0

        while curA:
            curA = curA.next
            la += 1
        
        while curB:
            curB = curB.next
            lb += 1
        
        d = abs(la-lb)
        lmax = la if la > lb else lb
        node1 = headA if la > lb else headB
        node2 = headA if la <= lb else headB
        for _ in range(d):
            node1=node1.next
        while node1 and node2:
            if node1 == node2:
                return node1
            node1 = node1.next
            node2 = node2.next

        return None
    
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:

        nodeA, nodeB = headA, headB
        flag = 0
        while nodeA != nodeB:
            if not nodeA: 
                nodeA = headB
                flag += 1
            elif nodeA:
                nodeA = nodeA.next
            if not nodeB: 
                nodeB = headA
                flag += 1
            elif nodeB:
                nodeB = nodeB.next
            if nodeA == nodeB:
                return nodeA

        return None if headA != headB else headA
````
