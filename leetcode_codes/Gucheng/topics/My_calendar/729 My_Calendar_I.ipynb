```python
# segment tree

class Node:
    def __init__(self, l, r, val=0, left=None, right=None): # l, r included
        self.l = l
        self.r = r
        self.val = val
        self.left = left
        self.right = right

class SegTree:

    def __init__(self, lst):
        self.root = self.build(0, len(lst)-1)
        self.lst = lst
    
    def build(self, l, r):

        if l> r:
            return None
        elif l == r:
            node = Node(l, r, val = self.lst[l])
            return node
        
        mid = l + (r-l)//2
        node = Node(l, r)
        node.left = self.build(l, mid)
        node.right = self.build(mid+1, r)
        node.val = node.left.val if node.left else 0 + node.right.val if node.right else 0
        
        return node
    
    def find(self, node, ql, qr):
        l, r = node.l, node.r
        mid = l + (r-l)//2

        if qr <= mid:
            return self.find(node.left, ql, qr)
        elif ql<mid< qr:
            return self.find(node.left, ql, mid) + self.find(node.leftm, mid+1, qr)
        elif ql >= mid:
            return self.find(node.right, ql, qr)


    

```
