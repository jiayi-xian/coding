#### pins and columns

Pinterest app screen is two columns of images (pins).
Each pin has a top and bottom index (top < bottom), and has a column represented by "L" or "R".
Given an unsorted list of non-overlapping pins like

`pins = [(top_ind,bottom_ind,column),...,]`
and a screen_length of screen_len
Return the maximum number of pins the user can possibly see (fully).
That is, complete this:


def get_max_pins(pins,screen_len):
    max_pins = 0
	...
    return max_pins
Example:

input:
  `pins = [(1,4,"L"),(2,3,"R"),(4,8,"R"),(6,9,"L")] `
  screen_len = 5
output: 2

```python
def solution(pins, L):

    lpins = [pin for pin in pins if pin[2]== "L"]
    rpins = [pin for pin in pins if pin[2]== "R"]
    cur_t = pins[0][0]
    cur_b = cur_t + L
    j0, j1, k0, k1= 0, 0, 0, 0
    max_cnt = 0
    lpt, rpt = 0, 0

    for i in range(len(pins)):
        cur_t = pins[i][0]
        cur_b = cur_t + L
        flag = pins[i][2]
        lpt += flag == "L"
        rpt += flag == "R"
        cur_cnt = 0
        

        if flag == "R":
            while j0<=len(lpins)-1 and lpins[j0][0] < cur_t: # and pins[j0][2] == "L" count L pins
                j0 += 1
            j1 = max(j1, j0)
        else:
            j0 = lpt -1
            j1 = j0
        while j1 <= len(lpins)-1 and lpins[j1][1] <= cur_b:
            j1 += 1
        
        cur_cnt += max(0, j1 - j0)

        if flag == "L": 
            while k0<=len(rpins)-1 and rpins[k0][0] < cur_t:
                k0 += 1
            k1 = max(k1, k0)
        else:
            k0 = rpt -1
            k1 = k0
        while k1 <= len(rpins)-1 and rpins[k1][1] <= cur_b:
            k1 += 1
        
        cur_cnt += max(0, k1 - k0)

        max_cnt = max(max_cnt, cur_cnt)
        print(cur_t, cur_b, j0, j1, k0, k1, cur_cnt)
    
    print(max_cnt)
    print(lpins, "\n", rpins)
    return max_cnt


pins = [(1,4,"L"),(2,3,"R"),(4,8,"R"),(6,9,"L")] 
screen_len = 5
pins = [(1,4,"L"),(2,3,"R"),(4,10,"R"),(5,6,"L")]
screen_len = 7
pins = [(1,4,"L"),(2,3,"R"),(4,15,"R"),(5,6,"L")]
screen_len = 7
pins = [(1,4,"L"),(1,1,"R"),(2,2,"R"),(3,3,"R"), (4,5, "L")]
screen_len = 2
solution(pins, screen_len)
```

```python
# Find maximum non overlapping intervals in a moving window
# 逻辑: sorted by start time 逐个区间看一遍 如果non overlap，加入， 如果overlap， 保留end time 小的那个(greedy)。 譬如说 [2,3], [2.5, 2.9]

def findMaxCoveredIntervals(intervals, L):
    # Step 1: Sort the intervals by start time
    intervals.sort(key=lambda x: x[0])

    max_covered = 0
    best_start = None
    best_end = None
    best_intervals = []

    # Step 3: Iterate through intervals to consider each as a window start
    for i in range(len(intervals)):
        current_start = intervals[i][0]
        current_end = current_start + L
        temp_intervals = []
        last_end = -1

        # Step 4: Find covered, non-overlapping intervals for the current window
        for j in range(i, len(intervals)):
            if intervals[j][0] >= current_start and intervals[j][1] <= current_end:
                if intervals[j][0] >= last_end:  # Step 5: Check for non-overlapping
                    temp_intervals.append(intervals[j])
                    last_end = intervals[j][1]

        # Step 6: Update maximum and optimal window if the current window is better
        if len(temp_intervals) > max_covered:
            max_covered = len(temp_intervals)
            best_start = current_start
            best_end = current_end
            best_intervals = temp_intervals

    # Step 7: Return the results
    return best_intervals, best_start, best_end

# Example usage
intervals = [[1, 4], [2, 5], [7, 9], [8, 10]]
L = 3
covered_intervals, window_start, window_end = findMaxCoveredIntervals(intervals, L)
print(f"Covered Intervals: {covered_intervals}")
print(f"Window Start: {window_start}, Window End: {window_end}")
```

```python
def shortestWay(source: str, target: str) -> int:
    # edge case: impossible
    
    i, cnt = 0, 1
    
    for c in target:
        
        i = source.find(c, i) 
        # words.find(char, idx) staring from idx (included) , 
        # find character char and return index, if not found, return -1
        # found
        
        if i!=-1:
            i += 1
            continue
        
        # not found
        if i == -1:
            
            i = source.find(c)
            
            if i == -1: return -1
            cnt += 1
            i += 1
    
    return cnt
```

#### find k nodes with depth d

```python
# Given a graph represented by a dictionary of integers, 
# a depth d as int, a number k as int, return k nodes with depth d


def find_k_nodes_at_depth(graph, d, k):
    # Initialize a queue for BFS with a tuple containing the node and its depth
    queue = deque([(node, 0) for node in graph.keys() if node not in graph.values()])
    visited = set()
    nodes_at_depth_d = []

    while queue and len(nodes_at_depth_d) < k:
        current_node, depth = queue.popleft()
        visited.add(current_node)

        if depth == d:
            nodes_at_depth_d.append(current_node)
        elif depth < d:
            for neighbor in graph[current_node]:
                if neighbor not in visited:
                    queue.append((neighbor, depth + 1))

    return nodes_at_depth_d[:k]
```

N agents, M customers, and T[] time for an agent to serve a customer. 
T[i] represents the amount of time it takes for an agent i to serve one customer. 
One agent can serve one customer at a time.

```python
# O(log(m*min(T))) minT because we could choose the fastest agenet to handle all the clients

def total_wait_time(n, m, times):
    # If there are more agents than customers, no wait time is needed
    if n > m:
        return 0

    # Initialize the search range for the minimum wait time
    min_time = 0
    max_time = max(times)
    low = min_time
    high = m * max_time

    # Binary search to find the minimum wait time that can serve all customers
    while low <= high:
        mid = low + (high - low) // 2
        sum = 0

        # Calculate the total number of customers that can be served in 'mid' time
        for time in times:
            sum += (mid // time)

        # If exactly 'm' customers can be served, 'mid' is a potential answer
        if sum == m:
            return mid

        # Adjust the search range based on the comparison
        if sum > m:
            high = mid - 1
        else:  # sum < m
            low = mid + 1

    # If no exact match is found, return -1
    return -1
```

#### Add operators

```python
class Solution:
    def addOperators(self, num: str, target: int) -> List[str]:
        L = len(num)
        ans = set()
        
        def backtrack(i, total, last, expr):
            if i == L:
                if total == target:
                    ans.add(expr)
                return
            
            for j in range(i, L):
                n = int(num[i:j+1])
                if i == 0:
                    backtrack(j+1, n, n, str(n))
                else:
                    backtrack(j+1, total + n, n, expr + '+' + str(n))
                    backtrack(j+1, total - n, -n, expr + '-' + str(n))
                    backtrack(j+1, total - last + last * n, last * n, expr + '*' + str(n))
                if n == 0:
                    break
                    
        backtrack(0, 0, 0, '')
        return list(ans)
```

#### Reservations (line sweaping)

```python
# reservations 
from collections import defaultdict
def solution(reservations, op, off, maxCap):
    """
    reservations: [(start, end, capacity)]
    """
    res = []
    ts = []
    cur = 0
    pt = op
    cap_D = defaultdict(int)

    def str2int(tt):
        tt_lst = tt.split(":")
        return int(tt_lst[0]+tt_lst[1])

    cap_D[op] = 0
    cap_D[off] = 0
    for start, end, cap in reservations:
        #start, end = str2int(start), str2int(end)
        cap_D[start] += cap
        cap_D[end] -= cap
        ts.append(start)
        ts.append(end)
    
    #op, off = str2int(op), str2int(off)
    ts.append(op)
    ts.append(off)
    ts = list(set(ts))# 不要用ts append 容易错 如果没有这一步 会出现重复的时间点 这样在下一步计算中就会出错
    ts = list(cap_D.keys())

    ts.sort(key = lambda x : str2int(x))

    for ct in ts:
        if cap_D[ct] != 0 or str2int(ct) == str2int(off):
            if str2int(ct) - str2int(pt) > 0:
                res.append([pt, ct, maxCap-cur])
            pt = ct
            cur += cap_D[ct]
    
    return res, cap_D, ts
        


import random
from datetime import datetime, timedelta

# Function to generate random reservation times and capacities

rs = [['02:00', '07:00', 78],['6:00', '15:00', 67],['12:00', '19:00', 89],
 ['17:00', '21:30', 87],
 ['21:00', '22:00', 63],
 ['22:00', '23:00', 93],
 
 
 ]
solution(rs, "2:00", "23:30", 500)

    


```
