  + 思路
    在第一个方法中，根据问题的提示，执行两轮 BFS（广度优先搜索）。
    对于那些不熟悉 BFS 概念的人，我们建议你阅读 BFS。
    在这个问题中，要求找到图的直径。根据定义，图的直径是两个距离最远的节点之间的距离。
    根据这个定义，如果我们能够找到距离所有节点最远的两个节点，问题就可以解决了。
    让我们将这两个节点称为极端外围节点。
    首先，可以肯定的是，从图中的任何节点开始，如果我们运行 BFS 遍历，我们最后访问的节点将是极端节点之一。
    作为极端外围节点，它应该是远离图中任何其他节点的节点之一。
    给定任何节点，从该节点开始的最长距离必须以一个极端外围节点结束。
    一旦我们确定了极端外围节点之一，然后我们可以再次应用 BFS 遍历。 先从已确定的极端外围节点开始。 在第二次 BFS 遍历结束时，我们将登陆在另一个极端外围节点上。根据定义，我们遍历的距离将是图的直径。
  + 算法
    基于上述思路，主要算法是通过 BFS 遍历来找到两个极端外围节点。
    定义一个名为 bfs(start) 的函数，该函数返回两个结果：**1）。**从start节点开始的最远节点；**2）。**起点与最远节点之间的距离。
    正如其名称所示，我们可以应用 BFS（广度优先搜索）遍历来实现上述函数。
    一旦实现了 bfs(start)，我们只需调用它两次即可解决问题。
    在第一次调用中，我们将获得一个极端外围节点。
    通过获得的外围节点，我们然后再次调用该函数，以获得另一个极端外围节点，更重要的是获得两个极端外围节点之间的距离。

````python
"""
103	Binary Tree Zigzag Level Order Traversal 
103. 二叉树的锯齿形层序遍历
给你二叉树的根节点 root ，返回其节点值的 锯齿形层序遍历 。（即先从左往右，再从右往左进行下一层遍历，以此类推，层与层之间交替进行）。
"""
class Solution:
    def zigzagLevelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root: #TODO 注意edge case
            return []
        q = [root]
        q = deque(q)
        res = []
        d = 0
        while q:
            level = []
            for _ in range(len(q)):
                node = q.popleft()
                level.append(node.val)
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
            if d%2:
                res.append(level[::-1])
            else:
                res.append(level)
            d += 1
        return res
````

class Solution:

    def allPossibleFBT(self, N: int) -> List[TreeNode]:
        res = []
        if N == 1:
            return [TreeNode(0)]
        # 结点个数必须是奇数
        if N % 2 == 0:
            return []
        
        # 左子树分配一个节点
        left_num = 1
        # 右子树可以分配到 N - 1 - 1 = N - 2 个节点
        right_num = N - 2
        
        while right_num > 0:
            # 递归构造左子树

            # 递归构造右子树
            right_tree = self.allPossibleFBT(right_num)

                for j in range(len(right_tree)):
                    root = TreeNode(0)
                    root.left = left_tree[i]
                    root.right = right_tree[j]
                    res.append(root)
            left_num += 2
            right_num -= 2
        
        return res

递归解析:
1. 终止条件:
    a. 当越过叶节点，则直接返回 null ；
    b. 当 root 等于 $p, q$, 则直接返回 root ;
2. 递推工作:
    a. 开启递归左子节点, 返回值记为 left ;
    b. 开启递归右子节点，返回值记为 right ;
3. 返回值：根据 left 和 right ，可展开为四种情况；
    a. 当 left 和 right 同时为空 : 说明 root 的左 / 右子树中都不包含 $p, q$, 返回 null ;
    b. 当 left 和 right 同时不为空 : 说明 $p, q$ 分列在 root 的异侧（分别在 左/右子树）, 因此 root 为最近公共祖先, 返回 root;
    c. 当 left 为空, right 不为空 : $p, q$ 都不在 root 的左子树中, 直接返回 right 。具体可分为两种情况:
        a. $p, q$ 其中一个在 root 的右子树 中, 此时 right 指向 $p$ （假设为 $p$ ）；
        b. $p, q$ 两节点都在 root 的右子树 中, 此时的 right 指向 最近公共祖先节点;
    d. 当 left 不为空, right 为空: 与情况 3. 同理;

````python
# 1644. Lowest Common Ancestor of a Binary Tree II (可能不存在)
"""
简单证明一下思路，如果某个点x是p和q的祖先结点，
情况1是x的左子树含有p和q或者右子树含有p和q，
第二种情况是x本身就是p或q之一，而自己的左右子树中含有另外一个需要的值
树形DP，
dfs的过程的返回值的含义是以root为头结点的树中是否含有p或者q。
dfs的是不断越来越深的遍历，ans更新的一定是祖先结点，随着深度加深最后更新的ans一定是最近公共祖先


"""
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def __init__(self):
        self.ans = None

    def lowestCommonAncestor(self, root, p, q):
        self.dfs(root, p, q)
        return self.ans

    def dfs(self, root, p, q):
        if root is None:
            return False
        
        l = self.dfs(root.left, p, q)
        r = self.dfs(root.right, p, q)
        
        # If left and right are True, it means p and q are found in different subtrees of root.
        if l and r:
            self.ans = root
            
        # If root is either p or q and one of the subtrees contains the other, root is the LCA.
        if (root.val == p.val or root.val == q.val) and (l or r):
            self.ans = root
            
        # Return True if left or right subtree contains p or q, or if root is p or q.
        return l or r or root.val == p.val or root.val == q.val
````
