```python
from typing import *
from collections import *
import heapq
from itertools import *

"""
买卖股票的最佳时机 III
给定一个数组，它的第 i 个元素是一支给定的股票在第 i 天的价格。
设计一个算法来计算你所能获取的最大利润。你最多可以完成 两笔 交易
注意：你不能同时参与多笔交易，你必须在再次购买前出售掉之前的股票
测试链接 : https://leetcode.cn/problems/best-time-to-buy-and-sell-stock-iii
"""


class Solution:
    def maxProfit2(self, prices):
        n = len(prices)
        # dp1[i] : 0...i范围上发生一次交易，不要求在i的时刻卖出，最大利润是多少 # !
        dp1 = [0] * n
        min_price = prices[0]
        for i in range(1, n): # best time to sell stock version I
            min_price = min(min_price, prices[i])
            dp1[i] = max(dp1[i - 1], prices[i] - min_price)
        # best[i] : 0...i范围上，所有的dp1[i]-prices[i]，最大值是多少
        # 这是数组的设置完全是为了替代最后for循环的枚举行为
        best = [0] * n
        best[0] = dp1[0] - prices[0]
        for i in range(1, n):
            best[i] = max(best[i - 1], dp1[i] - prices[i])
        # dp2[i] : 0...i范围上发生两次交易，并且第二次交易在i时刻卖出，最大利润是多少
        dp2 = [0] * n
        ans = 0
        for i in range(1, n):
            # 不需要枚举了
            # 因为，best[i]已经揭示了，0...i范围上，所有的dp1[i]-prices[i]，最大值是多少
            dp2[i] = best[i] + prices[i]
            ans = max(ans, dp2[i])
        return ans
    
    def maxProfit3(self, prices):
        n = len(prices)
        dp1 = [0] * n
        best = [0] * n
        best[0] = -prices[0]
        dp2 = [0] * n
        ans = 0
        min_price = prices[0]
        for i in range(1, n):
            min_price = min(min_price, prices[i])
            dp1[i] = max(dp1[i - 1], prices[i] - min_price)
            best[i] = max(best[i - 1], dp1[i] - prices[i])
            dp2[i] = best[i] + prices[i]
            ans = max(ans, dp2[i])
        return ans

    def maxProfit4(self, prices):
        dp1 = 0
        best = -prices[0]
        ans = 0
        min_price = prices[0]
        for i in range(1, len(prices)):
            min_price = min(min_price, prices[i])
            dp1 = max(dp1, prices[i] - min_price)
            best = max(best, dp1 - prices[i])
            ans = max(ans, best + prices[i])
        return ans

if __name__ == "__main__":
    sol = Solution()

    res = sol.func()
    print(res)
```

```python

from typing import *
from collections import *
import heapq
from itertools import *

"""
LeetCode 206: Reverse Linked List
(Follow up: A linked list can be reversed either iteratively or recursively. Could you implement both?)
#### 206 reverse linked nodes
1. pre, cur
2. nex 存储cur.next (下一时刻的cur)
3. reverse：update cur.next 为pre
4. move
"""


class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        pre, cur= None, head # 1
        while cur:
            nex = cur.next # 2
            cur.next = pre # 3
            pre, cur = cur, nex # 4
        
        return pre
    
    def recur_reverseList(self, head):
        if not head:
            return None
        nex_head = self.recur_reverseList(head.next)
        if nex_head:
            nex_head.next = head

        return head


if __name__ == "__main__":
    sol = Solution()

    res = sol.func()
    print(res)
```
